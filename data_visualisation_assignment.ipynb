```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = px.scatter_3d(df, x='X', y='Y', z='Z', title='3D Surface Plot')
fig.show()
```

```python
#BOKEH
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.sampledata.iris import flowers
from bokeh.plotting import figure, output_file,show
flowers
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
x = np.linspace(0, 10, 500)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2)
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#markers based on the 'sizes' and 'colors' columns
x=[1,2,3,4,5]
y=[5,6,5,5,3]
output_file("scatter.html")
p=figure(title="scatter plot",x_axis_label='x',y_axis_label='y')
p.scatter(x,y,size=[10,20,30,40,50],color=['red','green','blue','yellow','orange'])
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
p = figure(x_range=fruits, title="Fruit Counts")
p.vbar(x=fruits, top=counts, width=0.5)
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="orange", line_color="white", alpha=0.5)
show(p)
```

```python
#.5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap
from bokeh.layouts import gridplot
from bokeh.models import BasicTicker, PrintfTickFormatter
import numpy as np

# Sample heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
values = data_heatmap.flatten()
min_value, max_value = np.min(values), np.max(values)
mapper = LinearColorMapper(palette="Viridis256", low=min_value, high=max_value)

p_heatmap = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y", tools="hover", tooltips=[("value", "@image")], width=600, height=600)
p_heatmap.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=mapper)
color_bar = ColorBar(color_mapper=mapper, major_label_text_font_size="8pt", ticker=BasicTicker(desired_num_ticks=10),
                     formatter=PrintfTickFormatter(format="%0.2f"), label_standoff=6, border_line_color=None, location=(0, 0))

p_heatmap.add_layout(color_bar, 'right')
show(p_heatmap)
```
